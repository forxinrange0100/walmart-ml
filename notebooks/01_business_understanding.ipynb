# Business Understanding
Walmart is one of the leading retail chains in the US, between 2010 and 2012, Walmart collected weekly data from 45 stores, including special events, holidays, and economic conditions such as CPI and unemployment rates.

## Business Problem
Walmart is currently unable to accurately predict weekly sales for its 45 stores. This is critial for estimating expected revenue per store and supporting downstream operational task, such as inventory forecasting and stock optimization (assuming Walmart provides more detailed operational data)

The problem can be addressed using machine learning models. In fact, the previous approach already involved some form of machine learning, although its details and performance are not known. Additionally, there are models specifically designed for time series forecasting, which means that building a forecasting solution does not require highly complex or costly infrastructure.

## Sucess Criteria and Scope
- Reduce the forecasting error to improve planning accuracy and support better operational and financial decisions, even though the performance of the previous solution is unknow.
- Walmart provides historical data and is already using a machine learning pipeline, indicating that the problem is feasible.
- The minimun deliverable is an operational machine learning system that enables Walmart staff to analize and visualize the results of the weekly sales forecasts for each store.

## AI Pattern
This problem aligns with the [Predictive Analytics and Decision Support pattern](https://www.pmi.org/blog/seven-patterns-of-ai#:~:text=5.%20Predictive%20analytics%20%26%20decision%20support) which requires historical data to generate reliable future forecasts.

## Potential Pitfalls
Data drif may occur due to external conditions such as market changes, seasonality shift, so overreliance on the model without human judgment can cause poor decisions.
